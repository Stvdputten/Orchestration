```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame assuming your DataFrame is called df_2
# Uncomment and create your DataFrame accordingly
# df_2 = pd.DataFrame({
#     'Connections': [/*data*/],
#     'Requests': [/*data*/],
#     'Benchmark': ['sn', 'mm', 'hr', /*more data*/],
#     'Latency': [/*data*/]
# })

# Define the fontsize variable
fontsize = 15
sns.set_context("notebook", font_scale=1.3)

# Create a FacetGrid
g = sns.FacetGrid(df_2, col="Connections", row="Requests", hue="Benchmark",
                  margin_titles=True, height=6, aspect=.6)

# Map the bar plot to the grid
g.map(sns.barplot, "Benchmark", "Latency", order=["sn", "mm", "hr"])

# Customize the legend with correct font size
# g.add_legend(fontsize=13, title="Benchmarks")

# Add bar labels with custom font size
for axes in g.axes:
    for ax in axes:
        for container in ax.containers:
            ax.bar_label(container, label_type='edge', padding=18, fmt='%.2f', fontsize=fontsize)

# Set y-axis labels with custom font size
g.set_ylabels("Avg 99th Percentile Tail Latency (ms)", fontsize=fontsize)



# Adjust y-tick labels with custom font size
for ax in g.axes.flat:
    yticks = [int(tick) for tick in ax.get_yticks()]
    # print(yticks)
    ax.set_yticklabels(yticks, fontsize=fontsize)
    # ax.set_ylabels
    
# # Increase the font size of row titles (connections labels)
# for text in g.row_names.:
#     print(text)
#     # text.set_size(fontsize)
# g.axes[0,0].set_xlabel('axes label 1')
# g.axes[0,1].set_xlabel('axes label 2')
    
# Set x-axis labels and tick parameters for both x and y ticks
g.set_xlabels(fontsize=fontsize)
g.set_xticklabels(fontsize=fontsize)
    


# Save the figure
g.savefig("Figures/exp2-connections.png", bbox_inches='tight')
# g.savefig("Figures/exp2-connections.png")

# Display the plot
plt.show()
```
