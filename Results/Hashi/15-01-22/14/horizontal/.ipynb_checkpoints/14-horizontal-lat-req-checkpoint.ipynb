```python
import os
cwd = os.getcwd()
print(cwd)
directory = "../Figures"
path = os.path.join(cwd, directory) 
print(path)
try: 
    os.mkdir(path) 
except OSError as error: 
    print(error)  
```

```python
%env var=15
%env var_name=horizontal
%env exp=exp14

%env location=../Figures/
%env sn=sn
%env mm=mm
%env hr=hr
%env orch=nomad

!pwd
!echo

# what to copy
!cp ../$orch-$sn-* .
!cp ../$orch-$mm-* .
!cp ../$orch-$hr-* .

!for file in nomad-* ; do echo $file; done  
```

```python
# print which params to include
!for file in $orch-$sn-*; do echo $file | cut -d- -f $var ; done
!for file in $orch-$mm-*; do echo $file | cut -d- -f $var ; done
!for file in $orch-$hr-*; do echo $file | cut -d- -f $var ; done
```

```python
# rename to param(s)
!for file in $orch-$sn-*; do mv $file $sn-$(echo $file | cut -d- -f $var) ; done
!for file in $orch-$mm-*; do mv $file $mm-$(echo $file | cut -d- -f $var) ; done
!for file in $orch-$hr-*; do mv $file $hr-$(echo $file | cut -d- -f $var) ; done
```

```python
# Overview of all
!hdr-plot --output $(echo $location)$orch-$sn-$var_name-$exp.png --nobox $sn-* --legend_fontsize 20
!hdr-plot --output $(echo $location)$orch-$mm-$var_name-$exp.png --nobox $mm-* --legend_fontsize 20
!hdr-plot --output $(echo $location)$orch-$hr-$var_name-$exp.png --nobox $hr-* --legend_fontsize 20
```

```python
from IPython.display import display, Image

exp=os.environ['exp']
var_name=os.environ['var_name']
location=os.environ['location']
orch=os.environ['orch']

for bench in ['sn','mm','hr']:
    filename=f"{location}{orch}-{bench}-{var_name}-{exp}.png"
    display(Image(filename=filename, retina=True))
```
